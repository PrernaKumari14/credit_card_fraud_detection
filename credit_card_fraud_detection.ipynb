```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#distribution of legit transaction and fraudlent transaction
cerdit_card_data['Class'].value_counts()
```

0-> normal transaction
1-> fraudlent transaction

```python
#this dataset highly unbalance
#seperating the data for analysis
legit=cerdit_card_data[cerdit_card_data.Class==0]#entire legit data get stored here
fraud=cerdit_card_data[cerdit_card_data.Class==1]#entired fraud data get stored here
print(legit.shape)
print(fraud.shape)
```

under sampling


bulding a sample dataset containing similar distri. of normal and fraud transaction


```python
new_var=  legit_sample =legit.sample(n=217)
```

concatenating two Dataframe

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

splliting the data into features and targets


```python
X= new_dataset.drop('Class', axis=1)

Y=new_dataset['Class']
print(X)
```

split the data into trainig data and testing data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

Model_traning


```python
model=LogisticRegression()
```

```python
#traning the model with trainig data
model.fit(X_train , Y_train)
```

Model Evalution

accuracy score

```python
X_train_predection=model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predection,Y_train)
```

```python
print('traning data accuracy : ',training_data_accuracy)
```

```python
X_test_predection=model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_predection,Y_test)

```

```python
print('testing data accuracy : ',testing_data_accuracy)
```
